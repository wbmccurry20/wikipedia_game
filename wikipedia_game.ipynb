```python
import wikipedia 
import sys
#method to define path from start page to goal page
def path(info):
    if info['parent']:
        path(info['parent'])
        print(' ->', end='')
    print(info['title'], end='')
    
def start_page(select):
    #user input for start page
    page = 0
    while not page:
        user_enter = input(' %s start page: ' % select)
        page = wikipedia.page(user_enter)
    return page

def finish(select):
    #user inpurt for goal page 
    page = 0 
    while not page:
        user_enter = input(' %s Goal page: ' % select)
        page = wikipedia.page(user_enter)
    return page    
            
def wiki_game():
    #Start of Wikipedia game
    print('Wiki Game\n')
    start = start_page('Enter')
    goal_page = finish('Enter')
    #create dictionary to store all page titles
    i = {}
    i[start.title] = {'title': start.title, 'path': 0, 'parent': None}
    
    j = [i[start.title]]
    print('Finding path between start and finish page:')
    while j:
        new = j[0]
        j = j[1:]
        try:
            current = wikipedia.page(new['title'])
            print('\t%s' % current.title)
            for link in current.links:
                if link not in i:
                    i[link] = {
                    'title': link,
                    'path': new['path'] + 1,
                    'parent': new
                    }
                    
                    if link == goal_page.title:
                        print("goal page found:")
                        print("path: ", end='')
                        path(i[link])
                        print()
                        sys.exit()
                    j.append(i[link])
        #handle pageError when filtering through pages
        except wikipedia.exceptions.PageError as e:
            print('Skipping:' + new['title'], e)
        #handling disambiguationError 
        except wikipedia.exceptions.DisambiguationError as e:
            i[e.title]={
                'title': e.title,
                'path': new['path'] + 1,
                'parent': new
            }
            for option in e.options:
                if option not in i:
                    i[option] = {
                        'title': option,
                        'path': new['path'] + 1,
                        'parent': new
                       }
        # allows for user to exit program
        except KeyboardInterrupt:
                print('Exiting')
                sys.exit()
                
                
wiki_game()
```
